```python
import os
import pandas as pd
import shutil
from sklearn.model_selection import train_test_split

# === Paths ===
base_dir = r'D:\Computer Vision Dataset'
final_dir = os.path.join(base_dir, "finalcsv")
os.makedirs(final_dir, exist_ok=True)

# Create new folders for images
for folder in ["Train", "Valid", "Test"]:
    os.makedirs(os.path.join(final_dir, folder), exist_ok=True)

# === Load datasets ===
df_train = pd.read_csv(os.path.join(base_dir, "train.csv"))
df_test = pd.read_csv(os.path.join(base_dir, "test.csv"))
df_valid = pd.read_csv(os.path.join(base_dir, "valid.csv"))  # âœ… correct function

# Combine all into one DataFrame
df_all = pd.concat([df_train, df_test, df_valid], ignore_index=True)
df_all = df_all.sample(frac=1, random_state=42).reset_index(drop=True)  # Shuffle

# === Split 80% Train, 10% Valid, 10% Test ===
train_df, temp_df = train_test_split(df_all, test_size=0.2, random_state=42)
valid_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

# Save CSVs
train_df.to_csv(os.path.join(final_dir, "train_split.csv"), index=False)
valid_df.to_csv(os.path.join(final_dir, "valid_split.csv"), index=False)
test_df.to_csv(os.path.join(final_dir, "test_split.csv"), index=False)

print("âœ… CSVs saved successfully.")

# === Move images to final folders ===

# All original image folders
image_sources = [
    os.path.join(base_dir, "Train"),
    os.path.join(base_dir, "Test"),
    os.path.join(base_dir, "Valid")
]

# Function to move image by ID
def move_image(img_id, dest_folder):
    for src_folder in image_sources:
        src_path = os.path.join(src_folder, img_id + ".jpg")
        if os.path.exists(src_path):
            shutil.copy2(src_path, os.path.join(dest_folder, img_id + ".jpg"))
            return True
    return False

# Move images to final Train
print("ğŸ“¦ Moving Train images...")
moved = 0
for img_id in train_df["id"].astype(str):
    if move_image(img_id, os.path.join(final_dir, "Train")):
        moved += 1
print(f"âœ… Train images moved: {moved}")

# Move images to final Valid
print("ğŸ“¦ Moving Valid images...")
moved = 0
for img_id in valid_df["id"].astype(str):
    if move_image(img_id, os.path.join(final_dir, "Valid")):
        moved += 1
print(f"âœ… Valid images moved: {moved}")

# Move images to final Test
print("ğŸ“¦ Moving Test images...")
moved = 0
for img_id in test_df["id"].astype(str):
    if move_image(img_id, os.path.join(final_dir, "Test")):
        moved += 1
print(f"âœ… Test images moved: {moved}")
```

```python
# === Final Verification ===

# Count CSV entries
train_count = len(train_df)
valid_count = len(valid_df)
test_count = len(test_df)

# Count images in folders
train_images = len(os.listdir(os.path.join(final_dir, "Train")))
valid_images = len(os.listdir(os.path.join(final_dir, "Valid")))
test_images = len(os.listdir(os.path.join(final_dir, "Test")))

print("\nğŸ“Š Final Dataset Summary:")
print(f"ğŸ§  Train CSV rows: {train_count} | Images: {train_images}")
print(f"ğŸ” Valid CSV rows: {valid_count} | Images: {valid_images}")
print(f"ğŸ§ª Test  CSV rows: {test_count} | Images: {test_images}")
print(f"ğŸ“ Total rows: {train_count + valid_count + test_count}")
print(f"ğŸ–¼ï¸ Total images: {train_images + valid_images + test_images}")
```

# Data Cleaning

```python
import pandas as pd, re, os
from pathlib import Path
from tqdm import tqdm

# ---------- 1. Folders ----------
BASE_CSV = Path(r"D:\Computer Vision Dataset\finalcsv")
BASE_IMG = Path(r"D:\Computer Vision Dataset\finalcsv")   # Train / Valid / Test live here
OUT_DIR  = BASE_CSV / "clean"
OUT_DIR.mkdir(exist_ok=True)

splits  = ["train_split", "valid_split", "test_split"]
IMG_DIR = {
    "train_split": BASE_IMG / "Train",
    "valid_split": BASE_IMG / "Valid",
    "test_split" : BASE_IMG / "Test",
}

EXT       = ".jpg"      # .png if thatâ€™s what you use
ID_COL    = "id"        # image-id column in your CSV
TEXT_COL  = "text"      # report column in your CSV
DELETE_IMAGES = True    # False â†’ dry-run

# ---------- 2. Boiler-plate remover ----------
BOILER = re.compile(
    r"""^(
        (?:As|In)\s+compared\s+(?:to|with)\b |                 # â€œAs compared to â€¦â€
        Compared\s+(?:to|with)\b |                             # â€œCompared to â€¦â€
        Comparison\s+is\s+made\s+(?:with|to)\b |               # â€œComparison is made â€¦â€
        In\s+comparison\s+(?:with|to)\b |                      # â€œIn comparison with â€¦â€
        Since\s+the\s+prior\b |                                # â€œSince the prior â€¦â€
        With\s+the\s+earlier\s+study\b |                       # â€œWith the earlier study â€¦â€
        No\s+(?:relevant|significant)\s+change\s+(?:as\s+)?    # â€œNo significant change as compared â€¦â€
            compared\s+(?:with|to)\b |                         
        Compared\s+to\s+prior\b                                # â€œCompared to prior â€¦â€
    )[^.]*[.\s]*                                               # eat text to first period
    """,
    flags=re.I | re.VERBOSE,
)

def strip_boiler(text: str) -> str:
    """Remove comparison boiler-plate; leave the rest unchanged."""
    return BOILER.sub("", text).strip()

# ---------- 3. Process each split ----------
for split in splits:
    csv_path = BASE_CSV / f"{split}.csv"
    df       = pd.read_csv(csv_path)

    # strip boiler-plate
    df[TEXT_COL] = df[TEXT_COL].apply(strip_boiler)

    # keep rows that still have text
    keep_mask = df[TEXT_COL].str.strip().astype(bool)
    df_keep   = df[keep_mask].reset_index(drop=True)
    df_drop   = df[~keep_mask]

    # optionally delete orphaned images
    img_dir = IMG_DIR[split]
    for img_id in tqdm(df_drop[ID_COL], desc=f"Deleting images for {split}"):
        img_path = img_dir / f"{img_id}{EXT}"
        if img_path.exists():
            if DELETE_IMAGES:
                img_path.unlink()
            else:
                print("Would delete:", img_path)

    # save cleaned CSV
    out_path = OUT_DIR / f"{split}_clean.csv"
    df_keep.to_csv(out_path, index=False)
    print(f"{split}: saved â†’ {out_path}   (dropped {len(df_drop)} rows)")
```

```python
# --- loop through each CSV ---
for split in splits:
    csv_path = BASE_CSV / f"{split}.csv"
    df        = pd.read_csv(csv_path)

    # keep a copy of the original text
    orig_text = df[TEXT_COL].copy()

    # strip boiler-plate
    df[TEXT_COL] = df[TEXT_COL].apply(strip_boiler)

    # masks
    keep_mask    = df[TEXT_COL].str.strip().astype(bool)
    edited_mask  = (df[TEXT_COL] != orig_text)

    edited_kept  = (edited_mask & keep_mask).sum()
    dropped      = (~keep_mask).sum()
    kept_unchg   = (keep_mask & ~edited_mask).sum()

    # â”€â”€ delete orphaned images â”€â”€
    img_dir = IMG_DIR[split]
    for img_id in tqdm(df.loc[~keep_mask, ID_COL],
                       desc=f"Deleting images for {split}"):
        img_path = img_dir / f"{img_id}{EXT}"
        if img_path.exists() and DELETE_IMAGES:
            img_path.unlink()

    # â”€â”€ save cleaned CSV â”€â”€
    out_path = OUT_DIR / f"{split}_clean.csv"
    df.loc[keep_mask].to_csv(out_path, index=False)

    # â”€â”€ report â”€â”€
    print(
        f"{split}: edited & kept {edited_kept} rows â€¢ "
        f"dropped {dropped} rows â€¢ kept unchanged {kept_unchg} rows â†’ {out_path}"
    )
```
